##Time Series Analysis

### What is it?
Is a statistical method used to analyze data points collected and recorded over time. It's like looking at a sequence of measurements or observations taken at different points in time and trying to understand patterns or trends within that data.

Imagine you're tracking the temperature every day for a month. Time series analysis helps you answer questions like: How does the temperature change over time? Are there any repeating patterns, like warmer days on weekends? Can we predict future temperatures based on past data?

Time series analysis helps us understand and predict how data changes over time, making it useful for forecasting future trends or making decisions based on historical patterns.

##Components of TSA:

###Components that contribute to the overall behavior of a time series:

###Trend:

The trend component represents the long-term movement or directionality of the time series data. Trends can be upward (indicating growth), downward (indicating decline), or stable (indicating no significant change). Trends can provide valuable insights into underlying patterns or changes in the data over time.

###Seasonality:

Seasonality refers to periodic fluctuations or patterns in the data that occur at regular intervals, such as daily, weekly, monthly, or yearly cycles. These patterns are often influenced by seasonal factors like weather, holidays, or cultural events. Identifying and modeling seasonality is essential for understanding and forecasting time series data accurately.

###Cyclicality:

Cyclicality represents longer-term fluctuations in the data that are not strictly periodic like seasonality but occur over multiple time periods. Cycles may result from economic or business cycles, product life cycles, or other factors influencing the data's behavior. Unlike seasonality, cycles do not have fixed or predictable lengths.

###Irregularity (or Noise):

Irregularity refers to random fluctuations or noise in the data that cannot be attributed to the trend, seasonality, or cyclicality. These fluctuations may result from random variation, measurement errors, or other unpredictable factors. Irregular components can obscure underlying patterns in the data and make analysis and forecasting challenging.

###Level:

The level component represents the baseline or average value of the time series data over time, excluding the effects of trend, seasonality, and cyclicality. Understanding the level helps to assess the overall magnitude or scale of the data and is often used as a reference point for analyzing other components.




###By decomposing a time series into its constituent components (trend, seasonality, cyclicality, and irregularity), analysts can better understand its underlying structure and behavior. This understanding enables more accurate forecasting, anomaly detection, and decision-making based on time series data.

##Summary of the math used in TSA:


###Descriptive Statistics:
Before diving into more advanced techniques, it's essential to start with descriptive statistics. This includes calculating measures like the mean, median, variance, and standard deviation of the time series data. These statistics help summarize the central tendency, spread, and shape of the data distribution.

###Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF):
These functions are used to identify the correlation between a time series and its lagged values. ACF measures the correlation between observations at different time lags, while PACF measures the correlation between observations at a particular lag after removing the effects of shorter lags. These functions help identify the order of autoregressive (AR) and moving average (MA) terms in models like ARIMA.

###Autoregressive Integrated Moving Average (ARIMA) Models:
ARIMA models are a popular choice for modeling time series data. They combine autoregressive (AR), differencing (I), and moving average (MA) components to capture the underlying patterns in the data. The AR component represents the relationship between an observation and a linear combination of its past values, the I component deals with differencing to achieve stationarity, and the MA component represents the dependency between an observation and a residual error from a moving average model.

###Seasonal ARIMA (SARIMA) Models:
SARIMA models extend ARIMA models to account for seasonal patterns in the data. They incorporate seasonal differencing and seasonal AR and MA terms to capture periodic fluctuations that occur at fixed intervals.

###Exponential Smoothing Methods:
Exponential smoothing methods, such as simple exponential smoothing (SES), double exponential smoothing (DES), and triple exponential smoothing (TES or Holt-Winters method), are used to forecast future values of a time series by assigning exponentially decreasing weights to past observations. These methods are particularly useful for data with trend and/or seasonality.

###Vector Autoregression (VAR) Models:
VAR models are used when dealing with multiple time series variables that influence each other. They capture the dynamic interrelationships between variables by modeling them as linear functions of their lagged values.

###Machine Learning Techniques:
In addition to traditional statistical methods, machine learning techniques like neural networks, support vector machines (SVM), and random forests can also be applied to time series data for forecasting and anomaly detection.

##Univariate VS. Multivariate

###Univariate Time Series Analysis:

###Definition:

###Univariate time series analysis involves analyzing a single time-dependent variable or series over a period of time. It focuses on understanding the patterns, trends, and behavior of that single variable.

###Example:
Analyzing the historical stock prices of a single company over time to predict future price movements.

###Techniques:

Descriptive statistics, visualization (e.g., line plots, histograms), trend analysis, seasonality decomposition, and forecasting models (e.g., ARIMA, exponential smoothing).

###Multivariate Time Series Analysis:

###Definition:

Multivariate time series analysis involves analyzing multiple time-dependent variables or series simultaneously. It explores the relationships, dependencies, and interactions between these variables over time.

###Example:
Analyzing the simultaneous changes in temperature, humidity, and air pressure to forecast weather conditions.

###Techniques:

###Multivariate statistical methods such as multivariate regression, vector autoregression (VAR), structural VAR, and dynamic factor models. It also includes techniques for feature engineering, such as lagged variables and cross-correlation analysis.

###Key Differences:

###Focus:

###Univariate analysis focuses on a single variable, while multivariate analysis considers multiple variables.

###Complexity:

###Multivariate analysis is generally more complex due to the interrelationships between variables.

###Insight:

###Univariate analysis provides insights into the behavior of a single variable, while multivariate analysis allows for understanding the joint behavior and interactions between multiple variables.

###Applications:

Univariate analysis is common in forecasting single-variable time series data like sales, stock prices, or temperature. Multivariate analysis is used in scenarios where multiple variables influence the outcome, such as economic modeling, climate forecasting, and system monitoring.


##Limitations and considerations


###While time series analysis (TSA) is a powerful tool for understanding and predicting the behavior of sequential data, it also has several limitations:

###Stationarity Assumption:
Many time series models assume that the underlying data is stationary, meaning that its statistical properties such as mean, variance, and autocorrelation structure do not change over time. However, real-world data often exhibit non-stationarity, such as trends, seasonality, or structural breaks, which can violate the assumptions of traditional TSA models.

###Complexity and Interpretability:
Some time series models, especially those involving advanced techniques like machine learning algorithms or deep learning, can be complex and difficult to interpret. As a result, it may be challenging to explain the underlying factors driving the model's predictions or to gain actionable insights from the analysis.

###Data Quality and Missing Values:
Time series data can be susceptible to data quality issues such as missing values, outliers, or measurement errors. Dealing with such issues effectively is crucial for accurate analysis and forecasting but can be challenging, especially when data are noisy or incomplete.

###Limited Historical Data:
Time series analysis relies on historical data to identify patterns and make predictions about future behavior. However, in some cases, historical data may be limited or sparse, making it challenging to train accurate models, especially for long-term forecasting or in rapidly changing environments.

###Overfitting and Generalization:
Complex time series models with many parameters run the risk of overfitting, where the model learns to capture noise or random fluctuations in the training data rather than true underlying patterns. Overfit models may perform well on training data but generalize poorly to new or unseen data, leading to inaccurate predictions.

###Uncertainty and Risk:
Time series forecasts are inherently uncertain, and there is always a risk of prediction errors or unexpected deviations from the forecasted values. It's essential to quantify and communicate this uncertainty to stakeholders to make informed decisions and manage risk effectively.

###Assumption of Linearity:
Many traditional time series models, such as autoregressive integrated moving average (ARIMA), assume linear relationships between variables. However, real-world phenomena may exhibit nonlinear or complex behavior, which linear models may fail to capture adequately.

###Despite these limitations, time series analysis remains a valuable tool for analyzing and forecasting sequential data, especially when combined with domain knowledge, data preprocessing techniques, and careful model selection and evaluation. Awareness of these limitations can help analysts make informed decisions and mitigate potential pitfalls in their time series analysis workflows.

```python
# Import Necessary Libraries:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Load and Preprocess Data:
# Load the time series data
df = pd.read_excel('/content/drive/MyDrive/Copy of WeatherABQBP1.xlsx')

df['Datetime'] = pd.to_datetime(df['Datetime'])

# Set date column as index
df.set_index(df['Datetime'],inplace= True)
```

```python
def all_the_value_counts(df):
    """Prints the value counts for all columns in a pandas DataFrame

Counting values wide,
All columns, one by one revealed,
Insight to be found."""
    for column in df.columns:
        print(f"Value counts for {column}:")
        print(df[column].value_counts())
        print("\n")
```

```python
def convert_tuples_to_lists(df):
  """Loops through columns, converting tuples

From tuples to lists,
DataFrame transformed with ease,
Data now unbound."""
  for col in df.columns:
      if type(df[col][0]) == tuple:
          df[col] = df[col].apply(list)
  return df
```

```python
def find_tuple_cols(df):
  """Identifies tuples in all columns in a dataframe

Tuple columns sought,
Amidst the data's great breadth,
Found and returned whole."""
  tuple_cols = [col for col in df.columns if df[col].apply(lambda x: isinstance(x, tuple)).any()]
  return tuple_cols
```

```python
def remove_outliers(df):
  """Removes outliers from all columns in a dataframe with a threshold of 3* STD

Outliers expelled,
Data purged of impurities,
Clean and pristine now."""
  cleaned_data = df
  for col in df.columns:
      mean = df[col].mean()
      std = df[col].std()
      threshold = 3 * std
      lower_bound = mean - threshold
      upper_bound = mean + threshold
      cleaned_data[col] = df[col][(df[col] > lower_bound) & (df[col] < upper_bound)]
  return cleaned_data
```

```python
def count_the_cats(df):
    """Prints the value counts for categorical columns in a pandas DataFrame

Meow, count the cats,
Categorical values,
Pandas purrs content.
    """
    for column in df.select_dtypes(include=["category", "object"]).columns:
        print(f"Value counts for {column}:")
        print(df[column].value_counts())
        print("\n")
```

```python
df['Temperature'] = df['Temperature'].astype(str)

# Remove the 'degrees' symbol and any other non-numeric characters
df['Temperature'] = df['Temperature'].str.replace(' °F', ' ')

# Convert the 'temperature' column to a numeric data type (float or int)
df['Temperature'] = pd.to_numeric(df['Temperature'], errors='coerce')

# Display the DataFrame to verify the changes
print(df)
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size

# Plot the temperature column against another column (e.g., time)
plt.plot(df.index, df['Temperature'], marker='o', linestyle='-')

# Add labels and title
plt.xlabel('Datetime time in MST')
plt.ylabel('Temperature')
plt.title('Temperature Variation Over Time')

# Show grid
plt.grid(True)

# Show the plot
plt.show()
```

```python
decomposition = seasonal_decompose(df['Temperature'], model='additive', period=12)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
# Plot decomposition components
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(df.index, df['Temperature'], label='Original', color='blue')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(df.index, trend, label='Trend', color='red')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(df.index, seasonal, label='Seasonal', color='green')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(df.index, residual, label='Residual', color='brown')
plt.legend(loc='best')
plt.tight_layout()

# Step 5: Model Selection and Training (e.g., ARIMA)
# ARIMA, SARIMA, ETS, Prophet, etc.

# Step 6: Model Evaluation
# Evaluate the model using appropriate metrics

# Step 7: Forecasting
# Use the trained model to forecast future values
```

```python
# Train ARIMA Model for Forecasting:

# Fit ARIMA model
model = ARIMA(df['Temperature'], order=(5,1,0))  # Example order, you can tune this
fit_model = model.fit()

# Forecast next 12 time steps
forecast = fit_model.forecast(steps= 48)
```
